```python
#20
estudiantes = ['Ana', 'Luis', 'Carlos', 'María', 'Pedro']
notas = [18, 10, 14, 8, 16]

aprobados = []

# Recorremos las listas por su posición (0, 1, 2, 3, 4)
for i in range(len(estudiantes)):
    # Revisamos la nota del estudiante en la posición i
    if notas[i] >= 11:
        # Si aprobó, guardamos su nombre en la lista de aprobados
        aprobados.append(estudiantes[i])

print("Estudiantes aprobados:", aprobados)
# Resultado: ['Ana', 'Carlos', 'Pedro']
```
