```python

numero = float(input("introduce un numero: "))
if numero > 0:
    print("positivo")
elif numero < 0:
    print(" negativo")
else:
    print("cero.")

if numero % 2 == 0:
    print("es par")
else:
    print("es impar.")

if numero >= 18:
    print("ya podrías votar")
else:
    print("aún no podrías votar") 
```
