```python
total_cuenta = float(input("ingresa la cuenta total: "))
print(f"tu cuenta es: {total_cuenta}")

porcent_propina = float(input("ingresa propina deseada: "))
print(f"la propina deseada es: {porcent_propina}")

Monto_propina = total_cuenta * porcent_propina
print(f"El monto de la propina es: {Monto_propina}")
```
